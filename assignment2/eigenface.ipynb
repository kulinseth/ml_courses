```python
import numpy as np
from scipy import misc
from matplotlib import pylab as plt
import matplotlib.cm as cm
%matplotlib inline

import pandas as pd

data_directory = "/Users/spreeth/Documents/ModernAnalytics/Assignment_2/EigenFace/"


'''def pd_read_data(train_test):
    if(train_test=="train"):
        filenames = pd.read_table('faces/train.txt',header=0)
        digits= df['label'].values
        pixels=df.drop('label', axis=1).values
    
    return (pixels, digits)'''



def read_data(input_file):
    
    train_labels, train_data = [], []
    for line in open('faces/'+input_file):
        im = misc.imread(line.strip().split()[0])
        train_data.append(im.reshape(2500,))
        train_labels.append(line.strip().split()[1])
    
    train_data, train_labels = np.array(train_data, dtype=float), np.array(train_labels, dtype=int)
    plt.imshow(train_data[10, :].reshape(50,50), cmap = cm.Greys_r)
    plt.show()
    return train_data, train_labels

def mean_subtraction(data, avg_face):
    print("subtract mean face from every dace and print first face")
    subtracted_data = np.subtract(data, avg_face)
    plt.imshow(subtracted_data[0, :].reshape(50,50), cmap = cm.Greys_r)
    plt.show()

def average_face(data):
    print("printing average face")
    avg_face = np.array( map(np.average, np.transpose(data)))
    avg_face = avg_face.reshape(2500,)
    plt.imshow(avg_face.reshape(50,50), cmap = cm.Greys_r)
    plt.show()
    return avg_face

print("Reading train data..")
data, labels = read_data("train.txt")
avg_face = average_face(data)
mean_subtraction(data, avg_face)

print("Reading test data..")
test_data, test_labels=read_data("test.txt")
mean_subtraction(test_data, avg_face)
```
