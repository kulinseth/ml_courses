```python
def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(pd.Categorical(training_df['cuisine']).categories))
    plt.xticks(tick_marks, pd.Categorical(training_df['cuisine']).categories, rotation=45)
    plt.yticks(tick_marks, pd.Categorical(training_df['cuisine']).categories)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

from sklearn.metrics import confusion_matrix
actual_X_test = training_df[~X_train_msk]['cuisine']
df_diff = actual_X_test != predict_X_test
df_test = training_df[~X_train_msk]
df_test['pred'] = predict_X_test
df_test['dec_fn'] = X_test_decision_fn.tolist()
df_test[['id', 'cuisine', 'ingredients_string', 'pred']].to_csv('confusion.txt')
df_test[df_diff][['id', 'cuisine', 'ingredients_string', 'pred']].to_csv('diff.txt')
test_X_cm = confusion_matrix(actual_X_test, predict_X_test)
plt.figure(figsize=(10, 10))
plot_confusion_matrix(test_X_cm)
```
