4 Features/Attributes, 3 species with 50 samples of each species

```python
from matplotlib import pylab as plt
import numpy as np
import re

mypath = '/home/kulinseth/courses/modern-analytics/'
def species_color (s):
    if (s == 'Iris-setosa') :
        return 'r';
    if (s == 'Iris-versicolor') :
        return 'g';
    if (s == 'Iris-virginica') :
        return 'b';

def process_data(N, p):
    species = np.array([])
    dataset = np.array([])
    for line in open(mypath + 'iris.data'):
        attributes = map(float, line.split(',')[:4])
        species = np.append(species, species_color(line.rsplit(',', 1)[1].split()[0]))
        dataset = np.append(dataset, attributes)
    dataset = np.reshape(dataset, (N, p))
    return (dataset, species)

def plot (data, colors, p):
    #[xplots, yplots] = plt.subplots(p, p, sharex='col', sharey='row')
    for i in range(p):
        for j in range(p):
            if i != j :
                plt.scatter(data[:,i], data[:, j], c=species.tolist())
                plt.savefig(mypath + str(i) + str(j) + '.png')

if __name__ == "__main__":
    # The result is data in the format of N x p matrix
    # and species is in the color format which can be plotted
    N = 150
    p = 4
    (data, species) = process_data(N, p)
    plot(data, species, p)
```
